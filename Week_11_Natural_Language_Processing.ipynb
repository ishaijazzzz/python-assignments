```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
import pandas as pd

data = {
    "text": [
        "I love AI and deep learning",
        "Natural Language Processing is amazing",
        "I enjoy reading about machine learning",
        "Text preprocessing is important"
    ],
    "label": [1, 1, 1, 0]
}

df = pd.DataFrame(data)
df.head()

```

```python
import nltk
nltk.download('movie_reviews')
from nltk.corpus import movie_reviews

texts = [" ".join(movie_reviews.words(fileid)) for fileid in movie_reviews.fileids()]
labels = [1 if fileid.startswith('pos') else 0 for fileid in movie_reviews.fileids()]

df = pd.DataFrame({"text": texts, "label": labels})
df.head()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df['clean_text'] = df['text'].apply(clean_text)
```
