```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('creditcard_cleaned.csv')
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.hist(data['Amount'], bins=50, color='skyblue')
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='Class', data=data)
plt.title("Number of Fraud and Non-Fraud Transactions")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x='Class', y='Amount', data=data)
plt.title("Amount vs Transaction Class")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Amount")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data['Time'], data['Amount'], s=10, alpha=0.3)
plt.title("Amount vs Time of Transaction")
plt.xlabel("Time (in seconds)")
plt.ylabel("Amount")
plt.show()
```

```python
corr = data.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
